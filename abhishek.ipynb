```python
import pandas as pd
```

```python
df = pd.read_csv("dataset.csv", error_bad_lines=False)
```

```python
df = df.fillna(df.mean())
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
print(df.loc[0][1:-1])
```

```python
y = ["2004-05","2005-06","2006-07","2007-08","2008-09","2009-10","2010-11"]
```

```python
fig= plt.subplots(nrows=1, ncols=1, figsize=(15, 9))
plt.scatter(y,df.loc[0][1:-1],label="Mining", marker="o")
plt.scatter(y,df.loc[1][1:-1],label="Manufacturing", marker="^")
plt.scatter(y,df.loc[2][1:-1],label="Construction", marker="s")
plt.scatter(y,df.loc[3][1:-1],label="Total Industry", marker="x")
plt.ylabel("GCF in Industry (Rs crore)", fontsize= "12")
plt.xlabel("Years", fontsize= "12")
plt.title("Yearwise Gross Capital Formation (GCF) in Industry", fontsize= "15")
plt.legend()
```

```python
data = [df.loc[0][1:-1],df.loc[1][1:-1],df.loc[2][1:-1],df.loc[3][1:-1]]
      
fig, (ax1) = plt.subplots(nrows=1, ncols=1, figsize=(15, 9))
bplot1 = ax1.boxplot(data,vert=True,patch_artist=True, labels = ["Mining", "Manufacturing","Construction", "Total Industry"]) 
                     
                   
# Creating axes instance
colors = ['pink', 'lightblue', 'lightgreen', 'orange']

for patch,color in zip(bplot1['boxes'], colors):
    patch.set_facecolor(color)

# adding horizontal grid lines
for ax in [ax1]:
    ax.yaxis.grid(True)
    ax.set_xlabel('Industries')
    ax.set_ylabel('GCF in Industry (Rs crore)')
    ax.set_title("Yearwise Gross Capital Formation (GCF) in Industry", fontsize= "15")

plt.show()
# Creating plot
bp = ax.boxplot(data)

# show plot
plt.show()
```
